### 1. Import Libraries

### 2. Data preparation

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
data.tail(5)
```

```python
df = data
```

```python
df
```

```python
#Cleaning column names

df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Identifying the columns that need more cleaning
strings = list(df.dtypes[df.dtypes == "object"].index)
```

```python
strings
```

```python
#Changing the values to lower case and replacing spaces with underscores
for col in strings:
    df[col] = df[col].str.lower().str.replace(" ", "_")
```
